### train_log.csv

Modifications to meet the format of KDD cup 2015

```python
final_course_ids = df_course_final['course_id'].unique()
df_train_log_filtered = df_train_log[df_train_log['course_id'].isin(final_course_ids)]
df_train_log_filtered = df_train_log_filtered.reset_index(drop=True)
#print(df_train_log_filtered.head())
print(f"\nNumber of entries in filtered train_log: {len(df_train_log_filtered)}")
print(f"Number of unique courses in filtered train_log: {df_train_log_filtered['course_id'].nunique()}")
```

```python
df_train_log[df_train_log['object'].notna()]['action'].unique()
```

```python
action_mapping = {
    # Video actions
    'pause_video': 'video',
    'play_video': 'video',
    'seek_video': 'video',
    'stop_video': 'page_close',
    'load_video': 'video',
    
    'problem_get': 'access',
    
    'create_thread': 'discussion',
    'create_comment': 'discussion',
    'delete_thread': 'discussion',
    'delete_comment': 'discussion',
    
    'problem_check': 'problem',
    'problem_check_correct': 'problem',
    'problem_check_incorrect': 'problem',
    'problem_save': 'problem',
    'reset_problem': 'problem'
}

default_action = 'navigate'
```

```python
df_train_log_filtered.head()
```

```python
import numpy as np

df_train_log_filtered['action'] = df_train_log_filtered['action'].map(action_mapping).fillna(default_action)

# Display the first few rows to verify the mapping
df_train_log_filtered.loc[df_train_log_filtered['action'] == 'navigate', 'object'] = "3T6XwoiMKgol57cm29Rjy8FXVFcIomxl"
df_train_log_filtered.groupby('action').nunique()

```

```python
columns_to_drop = ['username', 'course_id', 'session_id']
df_train_log_modified = df_train_log_filtered.drop(columns=columns_to_drop)
df_train_log_modified = df_train_log_modified.rename(columns={'action': 'event','enroll_id':'enrollment_id'})
df_train_log_modified['source'] = np.where(df_train_log_modified['event'] == 'access', 'server', 'browser')

cols = df_train_log_modified.columns.tolist()
cols.insert(cols.index('event') , cols.pop(cols.index('source')))
df_train_log_modified = df_train_log_modified[cols]

# Display the first few rows of the modified DataFrame
print(df_train_log_modified.head())

df_train_log_modified.to_csv("data/raw_data/log_test.csv")
# Print some information about the modified DataFrame
print(f"\nTotal number of rows: {len(df_train_log_modified)}")
print(f"Columns in the modified DataFrame: {df_train_log_modified.columns.tolist()}")
print("\nDistribution of events:")
print(df_train_log_modified['event'].value_counts(normalize=True))
print("\nDistribution of sources:")
print(df_train_log_modified['source'].value_counts(normalize=True))
```

### Enrollment.csv

```python
enrollment_df = df_train_log_filtered[['enroll_id', 'username', 'course_id']].copy()
enrollment_df = enrollment_df.drop_duplicates()

enrollment_df = enrollment_df.sort_values(['enroll_id'])
enrollment_df = enrollment_df.rename(columns={'enroll_id':'enrollment_id'})
enrollment_df = enrollment_df.reset_index(drop=True)
enrollment_df.to_csv("data/raw_data/enrollment_test.csv")
```

### Truth.csv

```python
valid_enroll_ids = set(enrollment_df['enrollment_id'])
df_train_truth_filtered = df_train_truth[df_train_truth['enroll_id'].isin(valid_enroll_ids)]
df_train_truth_filtered = df_train_truth_filtered.reset_index(drop=True)
print(df_train_truth_filtered.head())
print(f"\nTotal number of rows in filtered truth data: {len(df_train_truth_filtered)}")
print(f"Number of unique enrollment IDs in filtered truth data: {df_train_truth_filtered['enroll_id'].nunique()}")
df_train_truth_filtered.to_csv("data/raw_data/test/truth_test.csv",header=False)
```
