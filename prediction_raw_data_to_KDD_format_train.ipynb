### course_info.csv 
- Modified it into the format of date.csv of KDD Cup and kept only those courses that are part of train_log.csv 

```python
import pandas as pd


course_times = df_train_log.groupby('course_id').agg({
    'time': ['min', 'max']
})

course_times.columns = ['start', 'end']
course_times = course_times.reset_index()
df_course_final = pd.merge(df_course[['course_id']], course_times, on='course_id', how='inner')
df_course_final['start'] = pd.to_datetime(df_course_final['start'])
df_course_final['end'] = pd.to_datetime(df_course_final['end'])

# Calculate the duration in weeks
df_course_final['duration_weeks'] = (df_course_final['end'] - df_course_final['start']).dt.days // 7

df_course_final = df_course_final[df_course_final['duration_weeks'] == 5]

# Now convert to date (after the duration calculation)
df_course_final['from'] = df_course_final['start'].dt.date
df_course_final['to'] = df_course_final['end'].dt.date

df_course_final = df_course_final.drop(columns=['duration_weeks','start','end'])

df_course_final = df_course_final.reset_index(drop=True)

# Display the first few rows of the result
print(df_course_final.head(),len(df_course_final))

df_course_final.to_csv('data/raw_data/date.csv',header=True)
```

```python
columns_to_drop = ['username', 'course_id', 'session_id']
df_train_log_modified = df_train_log_filtered.drop(columns=columns_to_drop)
df_train_log_modified = df_train_log_modified.rename(columns={'action': 'event','enroll_id':'enrollment_id'})
df_train_log_modified['source'] = np.where(df_train_log_modified['event'] == 'access', 'server', 'browser')

cols = df_train_log_modified.columns.tolist()
cols.insert(cols.index('event') , cols.pop(cols.index('source')))
df_train_log_modified = df_train_log_modified[cols]

# Display the first few rows of the modified DataFrame
print(df_train_log_modified.head())

df_train_log_modified.to_csv("data/raw_data/train/log_train.csv")
# Print some information about the modified DataFrame
print(f"\nTotal number of rows: {len(df_train_log_modified)}")
print(f"Columns in the modified DataFrame: {df_train_log_modified.columns.tolist()}")
print("\nDistribution of events:")
print(df_train_log_modified['event'].value_counts(normalize=True))
print("\nDistribution of sources:")
print(df_train_log_modified['source'].value_counts(normalize=True))
```
